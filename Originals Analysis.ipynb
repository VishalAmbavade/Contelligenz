```python
## Calculation of Average Completion Days
## There's no feature which cn tell if the user has watched the episode completely or left it in-between
## Furthermore, the completionRatio can tell the user has watchedd the series for more than one-episode in one sitting; 
## although, it is not sufficient 
```

```python
df_1['watchtime'] = df_1['watchtime'].dt.date
```

```python
df_1.head()
```

```python
df_4 = df_1.groupby(['userid', 'title_x']).agg({'episode_num':['min', 'max'], 'runtime':'first', 'duration':'first', 'watchtime':['min', 'max']})
```

```python
df_4.head(20)
```

```python
df_4.columns.get_level_values(1)
```

```python
df_4['completionDays'] = df_4['watchtime', 'max'] - df_4['watchtime', 'min'] + dt.timedelta(days = 1)
```

```python
df_4.tail()
```

```python
df_5 = df_1.groupby(['userid', 'title_x']).agg({'episode_num':['min', 'max'], 'runtime':'sum', 'duration':'sum', 'watchtime':['min', 'max']})
```

```python
df_5.head()
```

```python
df_5['completionRate'] = df_5['duration', 'sum'] / df_5['runtime', 'sum']
```

```python
df_5['completionDays'] = df_5['watchtime', 'max'] - df_5['watchtime', 'min'] + dt.timedelta(days = 1)
```

```python
df_5['avgSessionDuration'] = df_5['duration', 'sum'] / (df_5['runtime', 'sum'] * df_5['completionDays'].dt.days)
```

```python
df_5.head()
```

```python
### User don't tend to watch the whole show. Generally they leave it, as soon as credits start to appear. 
### That causes completion ration to be less than 1.0 which is considered ideal.

print(sum(df_5['completionRate'] > 0.90), sum(df_5['completionRate'] > 0.90) / len(df_5)) ## If we take 0.9 as the ideal completion rate
print(sum(df_5['completionRate'] > 0.85), sum(df_5['completionRate'] > 0.85) / len(df_5)) ## 0.85 as completion rate
print(sum(df_5['completionRate'] > 0.80), sum(df_5['completionRate'] > 0.80) / len(df_5)) ## 0.0 as completion rate
```

```python
df_5['watchedCompletely'] = df_5['completionRate'] > 0.85
```
